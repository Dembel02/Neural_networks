```python
import numpy as np # Подключаем библиотеку NumPy
from tensorflow.keras.models import Sequential # Подключаем класс создания модели Sequential
from tensorflow.keras.layers import Dense # Подключаем класс Dense - полносвязный слой
from tensorflow.keras.optimizers import Adam # Подключаем оптимизатор Adam
# from tensorflow.keras import utils
```

```python
model = Sequential()
model.add(Dense(2, input_dim = 2, use_bias=False, name='my_layers')) # Добавляем полносвязный слой с 2-мя нейронами
model.add(Dense(1, use_bias=False)) # Добавляем полносвязный слой с 1-м нейроном
```

```python
model = Sequential()
model.add(Dense(2, input_dim = 2, use_bias=False, name='my_layers')) # Добавляем полносвязный слой с 2-мя нейронами
model.add(Dense(1, use_bias=False)) # Добавляем полносвязный слой с 1-м нейроном
```

```python
model.summary() # Отобразим  данные созданной модели
```

```python
weights = model.get_weights() # Получим веса нашей модели (генерируются случайным образом)
print(weights)
```

```python
w1 = 0.42 # Зададим коэффициент w1 вручную
w2 = 0.15 # Зададим коэффициент w2 вручную
w3 = -0.56 # Зададим коэффициент w3 вручную
w4 = 0.83 # Зададим коэффициент w4 вручную
w5 = 0.93 # Зададим коэффициент w5 вручную
w6 = 0.02 # Зададим коэффициент w6 вручную
new_weights = [np.array([[w1,w3], [w2,w4]]), np.array([[w5],[w6]])] # сформируем список весов
print(new_weights) # отобразим сформированный список весов
model.set_weights(new_weights) # устанавливаем собственные веса модели
```

```python
weights = model.get_weights() # проверяем веса которые мы присвоили 
print(weights)
```

```python
x1 = 7.2 # Устанавливаем значение первого нейрона 
x2 = -5.8 # Устанавливаем значение второго нейрона
x_train = np.expand_dims(np.array([x1, x2]), 0) # Создадим набор данных для последующего обучения нейронной сети
x_train.shape
```

```python
k = np.array([x1, x2])
k.shape
```

```python
y_linear = model.predict(x_train) # Получим значение выхода сети, передав на вход вектор из двух элементов [x1, x2]
print(y_linear)
```

**Посчитаем выход сети вручную**

```python
H1 = x1 * w1 + x2 * w2 # Получим значение скрытого нейрона H1
H2 = x1 * w3 + x2 * w4 # Получим значение скрытого нейрона H2
print(H1) # Выводим значение скрытого нейрона H1
print(H2) # Выводим значение скрытого нейрона H2
```

```python
Y_linear = H1 * w5 + H2 * w6 # Посчитаем значение на выходе сети
print(Y_linear) # Выводим значение на выходе сети
```

```python
print('Значение модели посчитанной вручную ', round(Y_linear, 6))
print('Значение модели посчитанной нейронкой ', round(y_linear[0][0], 6))
```
